```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import pandas as pd

knowledge_capture_validation_data = pd.read_csv('KnowledgeCaptureValidationResults.csv',index_col=0)
knowledge_capture_validation_data.columns.name = 'Noise Ratio'

print(list(knowledge_capture_validation_data))

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_title("Knowledge capture validation results")
ax.set_xlabel('Rank')
ax.set_ylabel('Ratio of targeted genes')
for col in list(knowledge_capture_validation_data):
    if col != "ref":
        ax.plot(knowledge_capture_validation_data[col],knowledge_capture_validation_data.index,linewidth=1,label=col)
        ax.legend(loc='upper left')
    else:
        ax.plot(knowledge_capture_validation_data[col],knowledge_capture_validation_data.index,'k--',linewidth=3,label="(control)")
        ax.legend(loc='upper left',title="noise ratio")

    
#style = ['-']*knowledge_capture_validation_data.shape[1]
#style.append('--')
#lws = [1]*knowledge_capture_validation_data.shape[1]
#lws.append(2)
plt.savefig('test.png', bbox_inches='tight', pad_inches=0)
plt.close(fig)
```
