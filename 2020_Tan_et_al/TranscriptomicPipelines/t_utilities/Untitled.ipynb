```python
import pandas as pd
from enum import Enum

class GFF3Symbol(Enum):
    COMMENT = "#"
    EQUAL = "="
    ATTRIBUTESEPERATOR = ";"
    BLANK = " "
    
class GFF3Header(Enum):
    __order__ = 'SEQID SOURCE TYPE START END SCORE STRAND PHASE ATTRIBUTES'
    SEQID = "SEQID"
    SOURCE = "SOURCE"
    TYPE = "TYPE"
    START = "START"
    END = "END"
    SCORE = "SCORE"
    STRAND = "STRAND"
    PHASE = "PHASE"
    ATTRIBUTES = "ATTRIBUTES"
    
class GFF3AttributesHeader(Enum):
    NAME = "gene"
    GENESYNONYM = "gene_synonym"
    LOCUSTAG = "locus_tag"
    
data = pd.read_table('../../TestFiles/GCF_000001405.39_GRCh38.p13_genomic.gff', comment = GFF3Symbol.COMMENT.value, names = [e.name for e in GFF3Header])


def test_split(x, patterns):
    result = [""] * len(patterns)
            
    for s in x.split(GFF3Symbol.ATTRIBUTESEPERATOR.value):
        s2 = s.split(GFF3Symbol.EQUAL.value)
        for i in range(len(patterns)):
            if s2[0].replace(GFF3Symbol.BLANK.value,"") == patterns[i]: 
                result[i] = s2[1].replace(GFF3Symbol.BLANK.value,"")
                
    return ';'.join(result)

test_split_result = data["ATTRIBUTES"].apply(test_split, args=([e.value for e in GFF3AttributesHeader],))
print(test_split_result)

test_split_result2 = test_split_result.str.split(";",expand=True)
test_split_result2.columns = [e.value for e in GFF3AttributesHeader]
print(test_split_result2)
```
